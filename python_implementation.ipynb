```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import os
from os.path import isfile
```

```python
os.listdir()
```

```python
gps_dirs = [dir for dir in os.listdir() if 'GPS' in dir]
```

```python
gps_dirs
```

```python
list_std = []
for directory in gps_dirs:
    var_dir = (i for i in os.listdir(directory) if 'std_mean' in i) 
    list_std.append(np.genfromtxt(os.path.join(directory, *var_dir), delimiter=','))
len(list_std)
```

```python
##### Now Let's plot it
labels = ['ITU-GRACE16', 'ITU_GGC16', 'geco', 'eigen-6c4', 'egm2008']
# for i in range(len(list_std)):
plt.xlabel("Degrees")
plt.ylabel("Standard deviation \sigma")
for i in range(180, 2191 - 50, 200):
    plt.xlim([i, i+50])
    plt.ylim([0.35, .47])
    #plt.plot(list_std[0][:, 0], list_std[0][:, 1], label="ITU-GRACE16") 
    #plt.plot(list_std[1][:, 0], list_std[1][:, 1], label='ITU_GGC16')
    plt.plot(list_std[2][:, 0], list_std[2][:, 1], label="geco")
    plt.plot(list_std[3][:, 0], list_std[3][:, 1], label="eigen-6c4")
    plt.plot(list_std[4][:, 0], list_std[2][:, 1], label="egm2008")
    plt.legend(loc='best')
    plt.show()
    fname = 'figure%s' %(i + 200)
    plt.savefig(fname, dpi=1200)
```
